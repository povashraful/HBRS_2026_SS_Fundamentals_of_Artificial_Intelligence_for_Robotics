```python
start_position = [0,0]

final_goal = [5,5]

iteration = 0



while start_position != final_goal:
  user_input = input("Enter your move: (UP/DOWN/LEFT/RIGHT)\n ").strip().lower()
  if user_input not in ["up", "down", "left", "right"]:
        print("Wrong input, Please type (UP/DOWN/LEFT/RIGHT)")
        continue
  else:



    print("EPOCH----------------", iteration )

    up = [start_position[0], start_position[1]+1]
    down = [start_position[0], start_position[1]-1]

    right = [start_position[0]+1, start_position[1]]
    left = [start_position[0]-1, start_position[1]]

    print("CURRENT POSITION: ", start_position)
    print("FINAL GOAL: ", final_goal)

    if  user_input == "up":
      start_position = up
      print("NEW POSITION: ", start_position)


    elif  user_input == "down":
      start_position = down
      print("NEW POSITION: ", start_position)

    elif  user_input == "right":
      start_position = right
      print("NEW POSITION: ", start_position)

    elif  user_input == "left":
      start_position = left
      print("NEW POSITION: ", start_position)




    else:
      print("GOAL NOT REACHED")


    iteration += 1
```
