```python
import random

start_position = [0, 0]
final_goal = [5, 5]

max_iteration = 50
iteration = 0

grid_size = 6
print(grid_size)
path = [start_position.copy()]

while start_position != final_goal and iteration < max_iteration:

    user_input = random.choice(["up", "down", "left", "right"])

    print("EPOCH----------------", iteration)

    print("MOVE:", user_input)
    print("CURRENT POSITION:", start_position)
    print("FINAL GOAL:", final_goal)

    up = [start_position[0], start_position[1] + 1]
    down = [start_position[0], start_position[1] - 1]
    right = [start_position[0] + 1, start_position[1]]
    left = [start_position[0] - 1, start_position[1]]

    if user_input == "up":
        new_position = up


    elif user_input == "down":
        new_position = down

    elif user_input == "right":
        new_position = right

    elif user_input == "left":
        new_position = left

    if 0 <= new_position[0] < grid_size and 0 <= new_position[1] < grid_size:
        start_position = new_position
        print("NEW POSITION:", start_position)
    else:
        print("Out of boundary, staying at:", start_position)

    path.append(start_position.copy())
    iteration += 1

if start_position == final_goal:
        print("The walker has found the goal", path)
        print("The length of the path is: ", len(path))
else:
        print("Goal not found")
```

```python
import matplotlib.pyplot as plt

x = [p[0] for p in path]
y = [p[1] for p in path]

plt.figure(figsize=(8, 8))

plt.scatter(x, y)
plt.plot(x, y, alpha=0.5)


for i in range(len(path)):
    plt.text(x[i], y[i], str(i))

plt.scatter(0, 0, color="purple", label="Start")
plt.scatter(5, 5, color="red", label="Goal")

plt.title("Walker Path with Step Numbers")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [p[0] for p in path]
y = [p[1] for p in path]

plt.figure(figsize=(8, 8))

plt.scatter(x, y)
plt.plot(x, y, alpha=0.5)


visited_steps = {}

for step, position in enumerate(path):
    pos = tuple(position)

    if pos not in visited_steps:
        visited_steps[pos] = []

    visited_steps[pos].append(step)


for pos, steps in visited_steps.items():
    x_pos = pos[0]
    y_pos = pos[1]

    label = ",".join(str(s) for s in steps)

    plt.text(x_pos + 0.1, y_pos + 0.1, label, fontsize=8)

plt.scatter(0, 0, color="purple", label="Start")
plt.scatter(5, 5, color="red", label="Goal")

plt.title("Path to reach GOAL")
plt.legend()
plt.grid(True)
plt.show()
```
