```python

import random
nr = new_position = [[0,0]]
cr = current_position = [0,0]
rd = ["up", "down", "right", "left"]



choosing_random_movements = random.choice(rd)
choosing_random_movements

user_input = choosing_random_movements
print("OLD POSITION: ",current_position )
if user_input=="up":
  print("GO ",user_input)
  cr = [cr[0],cr[1]+1]
  nr = cr
elif user_input=="down":
  print("GO ",user_input)
  cr = [cr[0],cr[1]-1]
  nr = cr
elif user_input=="right":
  print("GO ",user_input)
  cr = [cr[0]+1,cr[1]]
  nr = cr
elif user_input=="left":
  print("GO ",user_input)
  cr = [cr[0]-1,cr[1]]
  nr = cr

print("NEW POSITION :", nr )
```
